### Google Engineer Prompt

```python
from langchain.messages import SystemMessage,HumanMessage
from langchain.agents import create_agent
from langchain_ollama import ChatOllama

llm = ChatOllama(model='mistral:latest',temperature=0.2)

prompt1 = """You are a senior software engineer at Google with 15 years of experience 
who has personally mentored over 50 junior developers. You design learning 
plans that are practical, project-driven, and mirror how real engineers 
learn on the job — not textbook theory. You know exactly which concepts 
beginners waste time on and which ones unlock everything else."""

agent = create_agent(llm,system_prompt=prompt1)

user_input = HumanMessage(content="how should i learn Python-OOPS")

messages = []
messages.append(user_input)

response = agent.invoke({"messages":messages})
```

```python
print(response['messages'][-1].content)
```

### logs
- valid-json: NO
- Schema match: NO
- progression-logic: Flat
- Resource quality : Generic
- Output Consistency:Different Structure

### Few Shot Prompting
#### Prompt Variant 0

```python
### Zero short prompting technique
from langchain_core.prompts import ChatPromptTemplate

human_instruction ="""Create a 5-day learning plan for the topic: {{topic}}
Learner level: {{level}}
Make each day progressively harder. Include 2-3 objectives, 
2 resources, and 2 tasks per day."""

chat_prompt_template = ChatPromptTemplate.from_messages([
  ('human',human_instruction)
])

chain = chat_prompt_template | llm 

user_input = {"topic":"python-OOPS","level":"low"}

response = chain.invoke({"messages":user_input})
```

```python
print(response.content)
```

### logs

### Prompt Variant 3

```python
### Negative shorts pompting technique - tell llm what not to do
from langchain_core.prompts import ChatPromptTemplate

human_instruction ="""Create a 5-day learning plan for:
Topic: {{topic}}
Level: {{level}}

STRICT RULES:
- Do NOT repeat the same resource across multiple days
- Do NOT make Day 2 and Day 3 the same difficulty
- Do NOT use vague tasks like "read about X" or "explore Y"
- Do NOT include duration_hours as a string — it must be a number
- Every task must be something the learner can physically DO and verify"""

## {{}} - langhchain thinks it literal
## {} - langchain thinks its variable

chat_prompt_template = ChatPromptTemplate.from_messages([
  ('human',human_instruction)
])

chain = chat_prompt_template | llm 

user_input = {"topic":"python-OOPS","level":"low"}

response = chain.invoke({"messages":user_input})
```

### COT - Variant 5 (COT + Role + Few shot)

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import JsonOutputParser

parser = JsonOutputParser()

human_instruction ="""You are a senior engineer mentor who designs practical learning plans.

Think through this before responding:
- What is the logical concept ladder for {topic} at {level}?
- Which day should introduce the hardest concept?
- What hands-on task proves the learner understood each day?

Here is an example of your output style for Day 1 of "Git Basics":
{{
  "day": 1,
  "title": "Understanding Version Control",
  "objectives": ["Understand why version control exists",
                 "Difference between Git and GitHub"],
  "resources": ["Git Docs - Getting Started", "Pro Git Book Ch.1"],
  "tasks": ["Make your first commit", "Push a repo to GitHub"],
  "duration_hours": 1.5
}}

Now output a complete 5-day JSON plan for:
Topic: {topic}
Level: {level}"""

## {{}} - langhchain thinks it literal
## {} - langchain thinks its variable

chat_prompt_template = ChatPromptTemplate.from_messages([
  ('human',human_instruction)
])

chain = chat_prompt_template | llm | parser

user_input = {"topic":"python-OOPS","level":"low"}

response = chain.invoke(user_input)
```

```python
print(response)
```

- valid-json: Yes
- Schema match: Yes
- progression-logic: Excellent

```python
from rich.console import Console
from rich import box
from rich.table import Table
def json_to_display_table(data:dict):
  console = Console(record =True)

  if isinstance(data,dict):
    table = Table(title='Planner-Agent-Response',title_style="green",box= box.ROUNDED,show_header=True, header_style="bright_magenta")
    table.add_column("Day", style="bold yellow", justify="center", width=15)
    table.add_column("Topic & Details", style="white", width=30)
    table.add_column("Objectives & Tasks", style="green", width=40)
    table.add_column("Resources", style="magenta", width=35)
    table.add_column("Time", style="bright_cyan", justify="right", width=15)

    for day_info in data["days"]:
        # Format Topic column with Title
        topic_cell = f"[bold text_box_theme]{day_info['title']}[/]"

        # Format Objectives and Tasks into a combined actionable checklist
        objectives_list = "\n".join([f"• {obj}" for obj in day_info["objectives"]])
        tasks_list = "\n".join([f"❑ {task}" for task in day_info["tasks"]])
        actions_cell = f"[bold underline]Objectives:[/]\n{objectives_list}\n\n[bold underline]Tasks:[/]\n{tasks_list}"

        # Format Resources into a list
        resources_cell = "\n".join([f"🔗 {res}" for res in day_info["resources"]])

        # Append the fully styled row to the table grid
        table.add_row(
            f"#{day_info['day']}",
            topic_cell,
            actions_cell,
            resources_cell,
            f"{day_info['duration_hours']} hrs"
        )
  console.print(table)
  console.save_html('planner_agent_response.svg')
```
