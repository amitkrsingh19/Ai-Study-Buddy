### Day 5 

### Task : Prompt engineer Mistral to produce a 5-day learning plan in structured JSON. Parse & display

### Imports

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain.messages import SystemMessage
from langchain_core.output_parsers import StrOutputParser
from langchain_ollama.chat_models import ChatOllama
import json
from dotenv import load_dotenv
import os
```

### load env && System Instruction

```python
GEMINI_KEY = os.getenv('GEMINI_API_KEY')
MODEL = 'mistral:latest'

system_instruction = """
              You are an expert curriculum designer. 
              Your ONLY output must be valid JSON — no prose, no markdown fences, no explanation.
              Respond with a JSON object matching this exact schema: 
              { "topic": string, "level": string, "days":
              [{ "day": int, "title": string, "objectives": [string],
              "resources": [string], "tasks": [string], "duration_hours": float } ] }
              """

human_instruction = """
Create a 5-day learning plan for the topic: {topic}
Learner level: {level}
Make each day progressively harder. Include 2-3 objectives, 2 resources, and 2 tasks per day.
"""
```

### llm Setup & chaining

```python
#llm = ChatGoogleGenerativeAI(model='gemini-flash-2.5',temperature=0.2, api_key=GEMINI_KEY)
llm = ChatOllama(model=MODEL,temperature = 0.2)

user_inputs = {"topic":"Supervised-learning in Machine learning","level":"medium"}

#system message as system prompt
sys_msg = SystemMessage(content=system_instruction)

chat_prompt_template = ChatPromptTemplate.from_messages(
  [sys_msg,
   ('human',human_instruction)]
)

parser = StrOutputParser()

chain = chat_prompt_template | llm | parser

response = chain.invoke(user_inputs)
```

### parse output as json

```python
import json 

json_response = json.loads(response)
print(json_response['days'][-1])
```

### Json to Dataframe

```python
import pandas as pd 
df = pd.DataFrame.from_dict(json_response['days'])
```
