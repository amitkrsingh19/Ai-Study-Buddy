### Imports

```python
import os
from dotenv import load_dotenv
from langchain_ollama import ChatOllama
from langchain.agents import create_agent
from langchain.tools import tool
from langchain.messages import AIMessage, HumanMessage,ToolMessage
from tavily import TavilyClient
from tavily.exceptions import BadRequestError
```

### Load env

```python
load_dotenv()
TAVILY_API_KEY = os.getenv('TAVILY_API_KEY')
```

### Prompts

```python

# prompt template for the LLM
system_instruction = """You are a helpful mathematical and research assistant. 

Follow these strict rules:
1. IF the user provides a math problem or expression, you MUST use the `calculator` tool to solve it. Do not try to do the math in your head.
2. IF the user asks for factual information or top lists, you MUST use the `search_web` tool.
3. IF the user simply greets you, says 'thank you', or makes small talk, you must respond with normal text and DO NOT output any JSON or tool calls.

Always return the final answer to the user after using a tool."""
```

### Tools

```python
@tool
def calculator(expression:str):
  """This is a calculator tool Use it when you see a mathematical Expression."""
  return str(eval(expression)) 

@tool
def web_search():
  """If User asks Anything specific which you dont know use this search tool to get the infromation from the web."""
  return str("web-answer")


tools = [calculator,web_search]
```

### llm & Agent

```python
llm = ChatOllama(model="mistral:latest",temperature=0.5)

agent = create_agent(llm,tools=tools,system_prompt=system_instruction)
```

### Invoke : Calculator

```python
messages = [] 
messages.append(HumanMessage(content="7 * 8 * 9 / 5 * 9 * 10"))

last_message = messages[-1]

response = agent.invoke({"messages":last_message})
```

### Response

```python
print(response['messages'][-1].content)
```

### Invoke : Web_search

```python
messages.append(HumanMessage(content ="find the best article on machine learning."))

last_message = messages[-1]

response = agent.invoke({"messages":last_message})
```
